```python
# Checking essential packages and versions
import importlib, subprocess

def check(name, cmd=None, import_as=None):
    print(f"{name}:", end=" ")
    try:
        if cmd:
            out = subprocess.run(cmd.split(), capture_output=True, text=True)
            print(out.stdout.splitlines()[0])
        else:
            mod = importlib.import_module(import_as or name)
            print("", getattr(mod, "__version__", "version unknown"))
    except Exception as e:
        print(f"Not found — {e}")

check("cmake", "cmake --version")
check("ninja", "ninja --version")

# Python packages
check("numpy")
check("tensorboard")
check("torch")
check("torchvision")
check("torchaudio")
check("detectron2")
```

```python
import os
import json
import shutil

folder1_img = "/content/drive/MyDrive/DataMorphAI/Line detection model/AnnotationsLDM30/10LDMAnn/images"
folder1_ann = "/content/drive/MyDrive/DataMorphAI/Line detection model/AnnotationsLDM30/10LDMAnn/result.json"

folder2_img = "/content/drive/MyDrive/DataMorphAI/Line detection model/AnnotationsLDM30/20LDMAnn/images"
folder2_ann = "/content/drive/MyDrive/DataMorphAI/Line detection model/AnnotationsLDM30/20LDMAnn/result.json"

output_dir = "/content/merged_dataset"
output_img_dir = os.path.join(output_dir, "images")
os.makedirs(output_img_dir, exist_ok=True)

with open(folder1_ann) as f:
    coco1 = json.load(f)
with open(folder2_ann) as f:
    coco2 = json.load(f)

def clean_file_names(coco):
    for img in coco["images"]:
        img["file_name"] = os.path.basename(img["file_name"])
    return coco

coco1 = clean_file_names(coco1)
coco2 = clean_file_names(coco2)

merged = {
    "info": coco1.get("info", {}),
    "licenses": coco1.get("licenses", []),
    "categories": coco1["categories"],
    "images": [],
    "annotations": [],
}

image_id_offset = 0
annotation_id_offset = 0

def add_data(coco, img_folder, image_offset, ann_offset):
    new_images = []
    new_annotations = []
    id_map = {}
    for img in coco["images"]:
        old_id = img["id"]
        new_id = old_id + image_offset
        id_map[old_id] = new_id
        img["id"] = new_id
        file_name = os.path.basename(img["file_name"])
        img["file_name"] = file_name
        new_images.append(img)
        src_path = os.path.join(img_folder, file_name)
        dst_path = os.path.join(output_img_dir, file_name)
        if not os.path.exists(dst_path):
            shutil.copy2(src_path, dst_path)
    for ann in coco["annotations"]:
        ann["id"] += ann_offset
        ann["image_id"] = id_map[ann["image_id"]]
        new_annotations.append(ann)
    return new_images, new_annotations

images1, anns1 = add_data(coco1, folder1_img, image_id_offset, annotation_id_offset)
image_id_offset += len(images1)
annotation_id_offset += len(anns1)

images2, anns2 = add_data(coco2, folder2_img, image_id_offset, annotation_id_offset)

merged["images"].extend(images1 + images2)
merged["annotations"].extend(anns1 + anns2)

os.makedirs(output_dir, exist_ok=True)
with open(os.path.join(output_dir, "combined_result.json"), "w") as f:
    json.dump(merged, f)

print("Merge completed successfully!")
print(f"Merged images: {output_img_dir}")
print(f"Merged annotations: {os.path.join(output_dir, 'combined_result.json')}")
```

```python
import os
from detectron2.engine import DefaultTrainer
from detectron2.config import get_cfg
from detectron2 import model_zoo

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))

cfg.DATASETS.TRAIN = ("line_detection_dataset_v2",)
cfg.DATASETS.TEST = ()
cfg.DATALOADER.NUM_WORKERS = 2

cfg.OUTPUT_DIR = "/content/line_detector_output"
os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)

cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")

cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 5e-5
cfg.SOLVER.MAX_ITER = 5000
cfg.SOLVER.STEPS = []
cfg.SOLVER.WARMUP_ITERS = 500
cfg.SOLVER.WARMUP_METHOD = "linear"
cfg.SOLVER.WARMUP_FACTOR = 0.001

cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 64
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1

cfg.INPUT.MIN_SIZE_TRAIN = (512,)
cfg.INPUT.MAX_SIZE_TRAIN = 768
cfg.INPUT.MIN_SIZE_TEST = 512
cfg.INPUT.MAX_SIZE_TEST = 768

cfg.MODEL.META_ARCHITECTURE = "GeneralizedRCNN"
cfg.MODEL.ROI_HEADS.NAME = "StandardROIHeads"

cfg.TEST.EVAL_PERIOD = 0

print("Config created")
```

```python
from PIL import Image
import os
from tqdm import tqdm

def verify_images(folder):
    bad_files = []
    for filename in tqdm(os.listdir(folder)):
        if filename.endswith(".png"):
            path = os.path.join(folder, filename)
            try:
                img = Image.open(path)
                img.verify()
            except Exception:
                bad_files.append(path)
                os.remove(path)
    return bad_files

bad = verify_images("/content/merged_dataset/images")
print("Bad images removed:", bad)
```

```python
from detectron2.engine import DefaultTrainer
from detectron2.utils.events import CommonMetricPrinter, JSONWriter
from detectron2.engine import hooks
import os

class CustomTrainer(DefaultTrainer):
    def build_hooks(self):
        hooks_list = super().build_hooks()

        hooks_list = [h for h in hooks_list if not isinstance(h, hooks.PeriodicWriter)]

        hooks_list.append(
            hooks.PeriodicWriter(
                [CommonMetricPrinter(), JSONWriter(os.path.join(self.cfg.OUTPUT_DIR, "metrics.json"))],
                period=20,
            )
        )
        return hooks_list

os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)

trainer = CustomTrainer(cfg)
trainer.resume_or_load(resume=True)
trainer.train()
```

```python
from google.colab import files
import cv2
import torch
import os
import zipfile
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.model_zoo import get_config_file
from detectron2 import model_zoo

uploaded = files.upload()
image_path = list(uploaded.keys())[0]

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1
cfg.MODEL.WEIGHTS = "/content/line_detector_output/model_final.pth"
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.2
predictor = DefaultPredictor(cfg)

image = cv2.imread(image_path)
outputs = predictor(image)
instances = outputs["instances"].to("cpu")
boxes = instances.pred_boxes.tensor.numpy()
sorted_boxes = boxes[boxes[:, 1].argsort()]

output_folder = "/content/cropped_lines"
os.makedirs(output_folder, exist_ok=True)

for i, box in enumerate(sorted_boxes):
    x1, y1, x2, y2 = map(int, box)
    cropped = image[y1:y2, x1:x2]
    cv2.imwrite(os.path.join(output_folder, f"line_{i+1:03}.png"), cropped)

zip_path = "/content/cropped_lines.zip"
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for root, _, files_list in os.walk(output_folder):
        for file in sorted(files_list):
            zipf.write(os.path.join(root, file), arcname=file)

files.download(zip_path)
```

```python
# Precision recall curve

from detectron2.evaluation import COCOEvaluator
import json
import os
import matplotlib.pyplot as plt
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval

ann_file = "/content/merged_dataset/combined_result_fixed.json"
res_file = os.path.join(cfg.OUTPUT_DIR, "coco_instances_results.json")

coco_gt = COCO(ann_file)
coco_dt = coco_gt.loadRes(res_file)

coco_eval = COCOeval(coco_gt, coco_dt, "bbox")
coco_eval.evaluate()
coco_eval.accumulate()

precisions = coco_eval.eval["precision"][0, :, 0, 0, 2]
recalls = coco_eval.params.recThrs

plt.figure(figsize=(6, 4))
plt.plot(recalls, precisions, marker='.')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (IoU=0.50)")
plt.grid()
plt.show()
```
